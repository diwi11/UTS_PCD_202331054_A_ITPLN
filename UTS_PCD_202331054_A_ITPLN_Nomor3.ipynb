```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'gambargelap.jpg'
image = cv2.imread(image_path)
```

```python
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
panjang = len(rgb)
lebar = len(rgb[0])
```

```python
r = rgb[:, :, 0]
g = rgb[:, :, 1]
b = rgb[:, :, 2]
```

```python
citra_gray = (0.2989 * r + 0.587 * g + 0.1141 * b).astype(np.uint8)

alpha = 1.0
beta = 80    

citra_cerah = np.clip(citra_gray + beta, 0, 255).astype(np.uint8)
citra_contras = np.clip(alpha * citra_gray, 0, 255).astype(np.uint8)
citra_hasil = np.clip(alpha * citra_gray + beta, 0, 255).astype(np.uint8)

plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(rgb)
plt.title("Gambar Asli")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(citra_gray, cmap='gray')
plt.title("Gambar Gray")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(citra_cerah, cmap='gray')
plt.title("Gray yang Dipercerah")
plt.axis("off")

plt.subplot(2, 3, 4)
plt.imshow(citra_contras, cmap='gray')
plt.title("Gray yang Diperkontras")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.imshow(citra_hasil, cmap='gray')
plt.title("Gray Dipercerah + Kontras")
plt.axis("off")

plt.tight_layout()
plt.show()
```
