```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('gambarbaru.jpg')
```

```python
if img is None:
    print("Error: Gambar tidak dapat dimuat. Periksa path gambar.")
else:
    blue_channel, green_channel, red_channel = cv2.split(img)
    
    plt.figure(figsize=(12, 10))
    
    plt.subplot(2, 2, 1)
    plt.title('CITRA KONTRAS', fontsize=14, fontweight='bold')
    
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('on')  
    
    # Tampilkan biru
    plt.subplot(2, 2, 2)
    plt.title('BIRU', fontsize=14, fontweight='bold')
    plt.imshow(blue_channel, cmap='gray')
   
    # Tampilkan merah
    plt.subplot(2, 2, 3)
    plt.title('MERAH', fontsize=14, fontweight='bold')
    plt.imshow(red_channel, cmap='gray')
    
    # Tampilkan hijau
    plt.subplot(2, 2, 4)
    plt.title('HIJAU', fontsize=14, fontweight='bold')
    plt.imshow(green_channel, cmap='gray')
    
    plt.tight_layout() 
    plt.show()
    
    plt.figure(figsize=(15, 5))
    
    # Histogram untuk merah
    plt.subplot(1, 3, 1)
    plt.title('Histogram Merah')
    plt.xlabel('Intensitas Piksel')
    plt.ylabel('Jumlah Piksel')
    plt.xlim([0, 256])
    plt.grid(alpha=0.3)
    plt.hist(red_channel.ravel(), 256, [0, 256], color='red', alpha=0.7)
    
    # Histogram untuk hijau
    plt.subplot(1, 3, 2)
    plt.title('Histogram Hijau')
    plt.xlabel('Intensitas Piksel')
    plt.ylabel('Jumlah Piksel')
    plt.xlim([0, 256])
    plt.grid(alpha=0.3)
    plt.hist(green_channel.ravel(), 256, [0, 256], color='green', alpha=0.7)
    
    # Histogram untuk biru
    plt.subplot(1, 3, 3)
    plt.title('Histogram Biru')
    plt.xlabel('Intensitas Piksel')
    plt.ylabel('Jumlah Piksel')
    plt.xlim([0, 256])
    plt.grid(alpha=0.3)
    plt.hist(blue_channel.ravel(), 256, [0, 256], color='blue', alpha=0.7)
    
    plt.tight_layout()
    plt.show()
```
